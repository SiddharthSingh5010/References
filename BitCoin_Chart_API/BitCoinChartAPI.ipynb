```python
import pandas as pd 
import matplotlib.pyplot as plt
from pycoingecko import CoinGeckoAPI
```

```python
cg=CoinGeckoAPI()
bitcoindata=cg.get_coin_market_chart_by_id(id='bitcoin',vs_currency='usd',days=30)
```

```python
bitcoin_pricedata=bitcoindata['prices']
```

```python
data=pd.DataFrame(bitcoin_pricedata,columns=['Timestamp','Price'])
```

```python
data.head()
```

```python
data['Date']=pd.to_datetime(data['Timestamp'],unit='ms')
```

```python
data
```

```python
data=data.drop('Timestamp',axis=1)
```

```python
data
```

```python
fig = plt.figure(figsize=(20, 5))
ax = fig.add_subplot(111)
ax.set_ylabel("Price")
ax.set_xlabel("Date")
ax.set_title("BITCOIN CHART")
ax.grid()
ax.plot(data['Date'],data['Price'])
plt.show()
```
