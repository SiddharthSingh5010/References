```python
#Using KNN Imputer to Fill Missing data in Dataframe
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/jbrownlee/Datasets/master/horse-colic.csv',na_values='?',header=None)
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
imputer=KNNImputer(n_neighbors=5)
```

```python
print(f"Total missing values in Dataframe : {df.isna().sum().sum()}")
```

```python
dfnew=pd.DataFrame(imputer.fit_transform(df))
```

```python
print(f"Total missing values in Dataframe after the KNN Imputing : {dfnew.isna().sum().sum()}")
```
