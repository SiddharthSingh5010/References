```python
import sqlite3
from sqlalchemy import create_engine
import pandas as pd
import numpy as np
```

```python
csv_database = create_engine('sqlite:///csv_database.db')
```

```python
chunk_size=50000
batch_no=1
for chunk in pd.read_csv('flights.csv',chunksize=chunk_size,iterator=True):
    chunk.to_sql('chunk_sql',csv_database, if_exists='append')
    batch_no+=1
    print('index: {}'.format(batch_no))
```
