```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#Simple Line Plot
x=np.arange(1,11)
y=np.arange(1,11)
plt.plot(x,y);
plt.title('My First Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
#Creating Two Line Plots in the same plot
x=np.arange(1,246)
y=np.arange(1,246)
plt.plot(x,y,color='green',linestyle=':',linewidth=5);
plt.plot(y,2*y)
plt.title('My First Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
```

```python
#Creating two line plots up and down
x=np.arange(1,246)
y=np.arange(1,246)
fig=plt.figure()
plt.subplot(2,1,1)
plt.plot(x,y,color='green',linestyle=':',linewidth=5);
plt.title('My First Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.subplot(212)
plt.plot(2*y,3*x,color='red')
plt.title('My Second Line Plot')
plt.xlabel('Index')
plt.ylabel('Price')
fig.tight_layout()
```

```python
#Creating Polar Plots side by side using subplot
fig=plt.figure()
x=np.arange(1,246)
y=np.arange(1,246)
plt.subplot(1,2,1,polar=True)
plt.plot(x,y,color='green',linestyle=':',linewidth=5);
plt.title('My First Line Plot')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.subplot(1,2,2,polar=True)
plt.plot(y,x)
plt.title('My Second Line Plot')
plt.xlabel('Index')
plt.ylabel('Price')
fig.tight_layout()
```

```python
#Creating Two Scatter Plot in the same plot
fig=plt.figure()
x=np.arange(1,21)
y=np.random.randn(20)
plt.scatter(x,y);
y=np.random.randn(20)
plt.scatter(x,y,marker='*',color='red');
```

```python
#Creating One Line and One Scatter Plot in the same plot
fig=plt.figure()
x=np.arange(1,21)
y=np.random.randn(20)
plt.plot(x,y);
y=np.random.randn(20)
plt.scatter(x,y,marker='*',color='red',s=100);
```

```python
#Creating Two Scatter Plot side by side
fig=plt.figure()
x=np.arange(1,21)
y=np.random.randn(20)
plt.subplot(121)
plt.scatter(x,y,s=10);
y=np.random.randn(20)
plt.subplot(122)
plt.scatter(x,y,marker='*',color='green',s=50)
fig.tight_layout()
```

```python
#simple bar plot
fig=plt.figure()
students={'Bob':80,'Sam':35,'Kat':67,'Mat':23,'Julia':93}
plt.bar(list(students.keys()),list(students.values()),color='orange');
plt.title('My First Bar Plot');
plt.xlabel('Students');
plt.ylabel('Marks');
```

```python
#Creating colorful Bar plot
fig=plt.figure()
students={'Bob':80,'Sam':35,'Kat':67,'Mat':23,'Julia':93}
plt.bar(x=list(students.keys()),height=list(students.values()),color=['red','orange','green','yellow','blue'],width=0.6);
plt.title('My First Bar Plot');
plt.xlabel(loc='left',xlabel='Students');
plt.ylabel(loc='bottom',ylabel='Marks');
```

```python
#Creating colorful Horizontal Bar plot
fig=plt.figure()
students={'Bob':80,'Sam':35,'Kat':67,'Mat':23,'Julia':93}
plt.subplot(211)
plt.barh(y=list(students.keys()),width=list(students.values()),color=['red','orange','green','yellow','blue'],height=0.6);
plt.title('My First Bar Plot');
plt.xlabel(loc='left',xlabel='Students');
plt.ylabel(loc='bottom',ylabel='Marks');
plt.subplot(212)
students={'Bob':80,'Sam':35,'Kat':67,'Mat':23,'Julia':93}
plt.bar(x=list(students.keys()),height=list(students.values()),color='orange',width=0.6);
plt.title('My First Bar Plot');
plt.grid(True)
plt.xlabel(loc='left',xlabel='Students');
plt.ylabel(loc='bottom',ylabel='Marks');
fig.tight_layout()
```

```python
#Stacked Bar Plot using matplotlib
x_axis=['A','B','C','D','E']
y1=[1,2,3,4,5]
y2=[1,3,5,2,1]
plt.figure(figsize=(8,4))
plt.bar(x_axis,y1,color='orange',label='Series 1');
plt.bar(x_axis,y2,bottom=y1,color='green',label='Series 2');
plt.legend();
```

```python
#Plotting simple Histogram
data=[1,3,3,3,3,4,4,5,9,9,6,8,8,8,8,7]
plt.hist(data,color='green')
plt.title('Histogram')
plt.xlabel('Data Distribution')
plt.ylabel('Counts');
```

```python
#Plotting simple step Histogram
data=[1,3,3,3,3,4,4,5,9,9,6,8,8,8,8,7]
plt.hist(data,color='green',histtype='step')
plt.title('Histogram')
plt.xlabel('Data Distribution')
plt.ylabel('Counts');
```

```python
#creating simple Box Plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
plt.boxplot(data);
```

```python
#creating simple Violin Plot
one=[1,2,3,4,5,6,7,8,9]
two=[1,2,3,4,5,4,3,2,1]
three=[6,7,8,9,8,7,6,5,4]
data=list([one,two,three])
plt.violinplot(data,showmeans=True);
```

```python
#Creating Simple Pie Chart
fruits=['Apple','Orange','Mango','Grapes']
values=[67,34,100,29]
plt.pie(values,labels=fruits);
```

```python
#Creating Pie Chart
fruits=['Apple','Orange','Mango','Grapes']
values=[67,34,100,29]
plt.pie(values,labels=fruits,autopct='%0.1f%%',shadow=True,radius=2);
```

```python
#Creating Pie Chart
fruits=['Apple','Orange','Mango','Grapes']
values=[67,34,100,29]
vegi=['Tomato','Potato','Onion']
vvalues=[20,50,50]
plt.pie(values,labels=fruits,autopct='%0.1f%%',shadow=True,radius=2);
plt.pie(vvalues,labels=vegi,autopct='%0.1f%%',radius=1);
```

```python
#Creating Donut Chart
fruits=['Apple','Orange','Mango','Grapes']
values=[67,34,100,29]
plt.pie(values,labels=fruits,autopct='%0.1f%%',shadow=True,radius=2);
plt.pie([1],colors='white',radius=0.8);
```
