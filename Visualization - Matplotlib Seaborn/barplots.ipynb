```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
%matplotlib inline
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
df=pd.read_csv('MARK_SCORED - Sheet3.csv')
```

```python
sns.countplot(df['Medium'],hue=df['Board']);
```

```python
df.columns
```

```python
df['total_marks']=df['English']+df['Language']+df['Maths']+df['Science']+df['SST']
```

```python
plt.pie(df['Board'].value_counts(),labels=df['Board'].value_counts().index,explode=(0.1,0,0,0),autopct="%0.2f%%",radius=3);
```

```python
sns.barplot('Medium','total_marks',data=df);
#By default sns.barplot gives Mean of values.
```

```python
sns.barplot('Medium','total_marks',data=df,ci=False);
#remove Confidence Interval on plot
```

```python
sns.barplot('Medium','total_marks',data=df,estimator=sum,ci=False);
#remove Confidence Interval on plot and plot Sum of all values not the means
```

```python
sns.barplot('Medium','total_marks',data=df,estimator=max,ci=False);
#remove Confidence Interval on plot and plot Maximum value of all values for each medium
```

```python
sns.barplot('Medium','total_marks',data=df,estimator=min,ci=False);
#remove Confidence Interval on plot and plot Minimum value of all values for each medium
```
