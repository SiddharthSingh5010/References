```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
from warnings import filterwarnings
filterwarnings('ignore')
sns.set_palette('deep')
```

```python
df=pd.read_csv('testsales.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date']=pd.to_datetime(df['Date'],format="%d-%m-%Y")
```

```python
df.head()
```

```python
df['Week']=df['Date'].apply(lambda x: x.week)
df['Month']=df['Date'].apply(lambda x: x.month)
df['Year']=df['Date'].apply(lambda x: x.year)
```

```python
df.head()
```

```python
df.drop([0,1,2],axis=0,inplace=True)
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.head(20)
```

```python
df=df.sort_values('Date').reset_index(drop=True)
```

```python
#Compare weekly
```

```python
df_week=df.groupby(['Year','Week']).sum()
```

```python
df_week.reset_index(inplace=True)
```

```python
df_week.head()
```

```python
df_week['Last Week Sales']=df_week.groupby('Year')['Sales'].shift(1)
```

```python
df_week
```

```python
df_week['Percentage Change']=((df_week['Sales']-df_week['Last Week Sales'])/df_week['Last Week Sales'])*100
```

```python
df_week.head()
```

```python
#Montly :
```

```python
df_month=df.groupby(['Year','Month']).sum()
```

```python
df_month.head()
```

```python
df_month['YTD_Month']=df_month.groupby('Year')['Sales'].cumsum()
```

```python
df_month
```

```python
df_month['Last Month Sales']=df_month.groupby('Year')['Sales'].shift(1)
df_month['Last 2 Month Sales']=df_month.groupby('Year')['Sales'].shift(2)
df_month['Last 3 Month Sales']=df_month.groupby('Year')['Sales'].shift(3)
```

```python
df_month.drop('Week',axis=1,inplace=True)
```

```python
df_month['Percentage Growth Last Month']=((df_month['Sales']-df_month['Last Month Sales'])/df_month['Last Month Sales'])*100
```
